```python
import rasterio as rio
import pandas as pd
import numpy as np

path = 'C:\\Users\\Luca\\Downloads\\GHS_BUILT_S_NRES_E2020_GLOBE_R2023A_54009_1000_V1_0\\GHS_BUILT_S_NRES_E2020_GLOBE_R2023A_54009_1000_V1_0.tif'
with rio.open(path) as src:
    
    # Se lee la imagen
    print('reading...')
    image = src.read()
    
    # Dimensiones de la imagen
    print('getting shapes...')
    bands, rows, cols = np.shape(image)
    
    # La imagen 2D se convierte a 1D
    print('getting radiances...')
    image1 = image.reshape (rows*cols,bands)
    
    
    # bounding box de la imagen
    print('getting bounds...')
    l,b,r,t = src.bounds
    
    # Resolution of image
    print('getting resolution...')
    res = src.res
    
    # Coordenadas X
    print('getting the coordinates...')
    X = np.arange(l, r, res[0])
    
    # Coordenadas Y
    Y = np.arange(t, b, -res[0])
```

```python
X,Y = np.meshgrid(X,Y)
```

```python
X = np.array(X.flatten())
Y = np.array(Y.flatten())
```

```python
df = np.column_stack((X, Y, image1))
```

```python
df = pd.DataFrame(df)
```

```python
df.columns = ['lon', 'lat', 'urban_builtup']
```

```python
df.replace(to_replace = 4294967295, value = np.nan, inplace = True)
df.dropna(inplace = True, ignore_index = True)
len(df)
```

```python
sav = "C:\\Users\\Luca\\Downloads\\RWI\\urban_builtup_bool"
bu = bu.rename(columns = {'urban_builtup': 'builtup'})

for file in os.listdir(countries):
    country = file[0:3]
    print(f'reading {country} population dataset...')
    pop = pd.read_csv(f'{countries}\\{file}')
    pop = pop[['quadkey', 'population']]
    print(f'population dataset read. len: {len(pop)}')

    print('merging with the builtup dataset...')
    data = pop.merge(bu, on = 'quadkey', how = 'left')
    print(data[['population', 'builtup']].isnull().sum())

    print(f'mean builtup for {country}: {data.builtup.mean()}')
    print(f'mean population density for {country}: {data.population.mean()}')
    print(f'number of areas with builtup over the average: {len(data[data.builtup >= data.builtup.mean()])}')
    print(f'number of areas considered urban: {len(data[data.population > data.population.mean()])}')
    data['urban_or_builtup'] = np.where((data.builtup >= data.builtup.mean()) | (data.population > data.population.mean()), 1, 0)
    print(f'percentage of areas considered urban or builtup: {100*data.urban_or_builtup.mean()}')
    data.to_csv(f'{sav}\\{country}_urban_builtup.csv', index = False)
```
