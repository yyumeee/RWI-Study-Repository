This short notebook illustrates the assignment of each quadkey 14 tile to its country, a crucial part of the study since all covariates are to be treated by their standardized score internal to their country.
It's important to note that there's multiple quadkey 14 tiles that overlap different countries, with some going as far as including 6-7 different countries. It's not a huge problem though, as we can just 'split' each tile across the different countries, with the same unstandardized value and then standardize separately.

```python
import pandas as pd
import os

with open("C:\\Users\\Luca\\Desktop\\Python\\Tirocinio\\RWI\\covariate\\quad_paesi2.txt", 'r') as file:
    list_quads = file.read()
    list_quads = list_quads.split(',')
```

We extract all the quadkey tiles from the population density datasets (as we explained in other notebooks, it's the method we used to establish which tiles are to be included in the study) and assign to each tile their country based on the dataset it was taken from.

```python
path = "C:\\Users\\Luca\\Downloads\\RWI\\pop_dens"

df = pd.DataFrame(columns = ['quadkey', 'country'])

for file in os.listdir(path):
    country = file[:3]

    temp = pd.read_csv(f'{path}\\{file}')
    print(f'{country} df: {temp.shape}')
    temp['country'] = country
    temp = temp[['quadkey', 'country']]

    df = pd.concat([df, temp], ignore_index = True)
    print(f'df: {df.shape}')
```
